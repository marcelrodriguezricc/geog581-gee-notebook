```python
# Libraries

import geemap, ee # Import geemap and earth engine together

print("Libraries imported")
```

```python
# Authenticate Google Earth Engine and initialize project

ee.Authenticate() # Authenticate Google Earth Engine account
ee.Initialize(project="geog-581-483717") # Initialize GEOG 581 project
print("Authenticated and initialized") # Print success
```

```python
# Load data and assign to variable (from Google Earth Engine Python tutorial)

jan_2023_climate = (
    ee.ImageCollection('ECMWF/ERA5_LAND/MONTHLY_AGGR')
    .filterDate('2023-01', '2023-02')
    .first()
)

jan_2023_climate

print("Data loaded")
```

```python
tempK = jan_2023_climate.select("temperature_2m") 

m = geemap.Map(center=[30, 0], zoom=2, ee_initialize=False) # Make sure to set ee_initialize = False

vis_params = {
    "min": 229,
    "max": 304,
    "palette": "inferno",
}

m.addLayer(tempK, vis_params, "Temperature (K)")
m
```
