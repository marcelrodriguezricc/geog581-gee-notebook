```python
# Libraries

import geemap, ee
import matplotlib.pyplot as plt

print("Libraries imported")

# Authenticate Google Earth Engine and initialize project

ee.Authenticate()
ee.Initialize(project="geog-581-483717")

print("Authenticated and initialized")
```

```python
# Load Landsat TOA image collection

landsat_08 = "LANDSAT/LC08/C02/T1_TOA"

try:
    landsat = ee.ImageCollection(landsat_08)
except:
    print(f"Failed to load dataset {landsat_08}")
else:
    print(f"Successfully loaded {landsat_08}")
```

```python
# Set a region of interest over Corvallis, OR

corvallis = ee.Geometry.Point([-123.2516667, 44.5688889])
```

```python
# Filter for image with lowest cloud cover in timespan

image = landsat.filterBounds(corvallis).filterDate('2022-01-01', '2022-12-30').sort('CLOUD_COVER').first()
```

```python
# Isolate RGB bands and use reducer to get min for each

rgb = image.select(['B4', 'B3', 'B2'])

rgb_min = rgb.reduceRegion(
    reducer=ee.Reducer.min(),
    geometry=image.geometry(),
    scale=30,
    maxPixels=1e9,
    bestEffort=True,
)

print(rgb_min.getInfo())
```
