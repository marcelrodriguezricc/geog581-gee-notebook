```python
# Libraries

import geemap, ee # Import geemap and earth engine together

print("Libraries imported")

# Authenticate Google Earth Engine and initialize project

ee.Authenticate() # Authenticate Google Earth Engine account
ee.Initialize(project="geog-581-483717") # Initialize GEOG 581 project

print("Authenticated and initialized")
```

```python
L8 = ee.ImageCollection("LANDSAT/LC08/C02/T1_TOA")

roi = ee.Geometry.Point(-123.02, 42.37)

filtered = L8.filterDate('2015-09-01', '2015-10-30').filterBounds(roi).filterMetadata("CLOUD_COVER",'less_than',10)

print(filtered)
```

```python
image = ee.Image(filtered.first())
rgb_vis = {"min":0, "max":0.3, "bands":['B4','B3','B2']}

Map = geemap.Map(ee_initialize=False)
Map.addLayer(image,rgb_vis, 'RGB')

ndvi = image.normalizedDifference(['B5', 'B4'])

ndvi_vis = {"palette":['FFFFFF','CE7E45','DF923D','F1B555','FCD163','99B718','74A901','66A000','529400','3E8601','207401','056201','004C00','023B01','012E01','011D01','011301'], "min": 0, "max": 0.7}

Map.addLayer(ndvi, ndvi_vis, 'NDVI')

Map.centerObject(ee.Geometry(roi), 8)

Map
```
