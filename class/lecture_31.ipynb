```python
# Libraries

import geemap, ee # Import geemap and earth engine together

print("Libraries imported")

# Authenticate Google Earth Engine and initialize project

ee.Authenticate() # Authenticate Google Earth Engine account
ee.Initialize(project="geog-581-483717") # Initialize GEOG 581 project

print("Authenticated and initialized")
```

```python
modis_ndvi = ee.ImageCollection('MODIS/MCD43A4_006_NDVI').filterDate('2022-05-01','2022-07-31')
modis_evi = ee.ImageCollection('MODIS/MCD43A4_006_EVI').filterDate('2022-05-01','2022-07-31')
```

```python
vegVisParams = {"min": 0, "max": 1, "palette": ['FFFFFF', 'CE7E45', 'DF923D', 'F1B555', 'FCD163', '99B718', '74A901', '66A000', '529400', '3E8601', '207401', '056201', '004C00', '023B01', '012E01', '011D01', '011301']}

Map = geemap.Map(ee_initialize=False)

Map.addLayer(modis_ndvi,vegVisParams,'MODIS NDVI')

Map
```

```python
map2 = geemap.Map(ee_initialize=False)
map2.addLayer(modis_evi , vegVisParams, "MODIS EVI")
map2
```

```python
import ipywidgets as widgets
from ipyleaflet import SplitMapControl


# Reduce collections to images for visualization
ndvi_img = modis_ndvi.mean()
evi_img  = modis_evi.mean()

# Make tile layers
left_layer  = geemap.ee_tile_layer(ndvi_img, vegVisParams, "MODIS NDVI")
right_layer = geemap.ee_tile_layer(evi_img,  vegVisParams, "MODIS EVI")

# Create split map
m = geemap.Map(ee_initialize=False, zoom=5)
m.add_control(SplitMapControl(left_layer=left_layer, right_layer=right_layer))
m
```
