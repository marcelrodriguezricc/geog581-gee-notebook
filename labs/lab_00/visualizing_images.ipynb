```python
# Libraries

import geemap, ee # Import geemap and earth engine together
from IPython.display import Image, display

print("Libraries imported")
```

```python
# Authenticate Google Earth Engine and initialize project

ee.Authenticate() # Authenticate Google Earth Engine account
ee.Initialize(project="geog-581-483717") # Initialize GEOG 581 project

print("Authenticated and initialized")
```

```python
# Instantiate an image
image = ee.Image("CGIAR/SRTM90_V4")

# Create the interactive map and "setCenter"
layer1 = geemap.Map(center=[36.2841, -112.8598], zoom=9, ee_initialize=False)

# Add the layer
layer1.addLayer(image, {}, "SRTM DEM") 
layer1
```

```python
# Print metadata
print('SRTM image', image)
```

```python
# Add a new layer with the grayscale range redefined
vis = {
    "min": 0,
    "max": 3000,
}

layer2 = geemap.Map(center=[36.2841, -112.8598], zoom=9, ee_initialize=False)
layer2.addLayer(image, {"min": 0, "max": 3000}, "custom visualization")
layer2
```

```python
# Add another layer, this time with a color palette
vis = {
    "min": 0,
    "max": 3000,
    "palette": ["blue", "green", "red"],
}
layer3 = geemap.Map(center=[36.2841, -112.8598], zoom=9, ee_initialize=False)
layer3.addLayer(image, vis, "custom palette")
layer3
```
