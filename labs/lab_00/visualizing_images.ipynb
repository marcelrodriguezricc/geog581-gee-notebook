```python
# Libraries

import geemap, ee # Import geemap and earth engine together

print("Libraries imported")
```

```python
# Authenticate Google Earth Engine and initialize project

ee.Authenticate() # Authenticate Google Earth Engine account
ee.Initialize(project="geog-581-483717") # Initialize GEOG 581 project

print("Authenticated and initialized")
```

```python
# Instantiate an image (same asset as the JS example)
image = ee.Image("CGIAR/SRTM90_V4")

# Create the interactive map and "setCenter"
m = geemap.Map(center=[36.2841, -112.8598], zoom=9, ee_initialize=False)

# Add the layer
m.addLayer(image, {}, "SRTM DEM") 
m
```

```python
# Print metadata
print('SRTM image', image)
```

```python
# Add a new layer with the grayscale range redefined
vis = {
    "min": 0,
    "max": 3000,
}
m.addLayer(image, {"min": 0, "max": 3000}, "custom visualization")
m
```

```python
# Add another layer, this time with a color palette
vis = {
    "min": 0,
    "max": 3000,
    "palette": ["blue", "green", "red"],
}
m.addLayer(image, vis, "custom palette")
m
```
